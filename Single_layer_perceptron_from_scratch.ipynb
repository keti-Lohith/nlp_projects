```python
class Perceptron:
    def __init__(self, no_features, eta):
        self.no_features = no_features
        self.eta = eta
        self.wts = numpy.zeros(1 + self.no_features)

    def activation(self, v):
        return 1 if v >= 0 else 0

    def prepend_bias(self, x):
        return numpy.insert(x, 0, 1)

    def predict(self, X):
        xi_bias = self.prepend_bias(X)
        v = self.wts.T.dot(xi_bias)# W^T *x
        y = self.activation(v)
        return y

    def train(self, X, D):
        for xi, d in zip(X, D):
            xi_bias = self.prepend_bias(xi)
            y = self.predict(xi)
            error = d - y
            delta_w = self.eta * error * xi_bias

            self.wts = self.wts + delta_w
```

```python
class Single_layer_perceptron:

    def __init__(self, no_features, no_classes, eta):
        self.no_classes = no_classes
        self.perceptrons = [Perceptron(no_features, eta) for _ in range(no_classes)]

    def train(self, X, D, epochs: int = 1):
        for _ in range(epochs):

            X_shuff, D_shuff = shuffle(X, D, random_state=0)

            for i in range(self.no_classes):
                self.perceptrons[i].train(X_shuff, D_shuff[:, i])

    def predict(self, x):
        return [percep.predict(x) for percep in self.perceptrons]

    def test(self, X):
        Y_pred = []
        for xi in X:
            y = self.predict(xi)
            Y_pred.append(y)
        return numpy.array(Y_pred)
```

```python
X=data.iloc[:,:-1]
y=data.iloc[:,-1]
r= 0.3
k = int(X.shape[0] *(1- r))
train_x, train_y = X[:k], dummies[:k]
test_x, test_y = X[k:],dummies[k:]
ETA = 0.9
EPOCHS = 100
```

```python
s = Single_layer_perceptron(train_x.shape[1], train_y.shape[1], ETA)
s.train(train_x.to_numpy(), train_y.to_numpy(), EPOCHS)
```

```python
Y_pred = s.test(test_x.to_numpy())
actual_labels = test_y.to_numpy()
predicted_labels =  Y_pred
accuracy =(accuracy_score(actual_labels, predicted_labels)*100)

print(f'Accuracy: {accuracy:.2f}')
```
