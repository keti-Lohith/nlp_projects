```python
class Perceptron:
    def __init__(self, no_features, eta):
        self.no_features = no_features
        self.eta = eta
        self.wts = numpy.zeros(1 + self.no_features)
    def sigmoid(x):
    return 1 / (1 + math.exp(-x))

    def activation(self, v):
        result = sigmoid(v)
        return result

    def prepend_bias(self, x):
        return numpy.insert(x, 0, 1)

    def wixi(self, X):
        xi_bias = self.prepend_bias(X)
        v = self.wts.T.dot(xi_bias)# W^T *x
        return v

    def train(self, X):
        for xi in X:
            v = self.wixi(xi)
            y = self.activation(v)
```

```python
import numpy as np

class NeuralNetwork:
    def __init__(self, input_size, hidden_size, output_size):
        # Initialize network architecture and random weights
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size

        # Initialize weights and biases for the hidden layer
        self.weights_hidden = np.random.randn(self.input_size, self.hidden_size)
        self.bias_hidden = np.zeros((1, self.hidden_size))

        # Initialize weights and biases for the output layer
        self.weights_output = np.random.randn(self.hidden_size, self.output_size)
        self.bias_output = np.zeros((1, self.output_size))

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def sigmoid_derivative(self, x):
        return x * (1 - x)

    def feedforward(self, x):
        # Calculate the hidden layer output
        self.hidden_layer_input = np.dot(x, self.weights_hidden) + self.bias_hidden
        self.hidden_layer_output = self.sigmoid(self.hidden_layer_input)

        # Calculate the output layer output
        self.output_layer_input = np.dot(self.hidden_layer_output, self.weights_output) + self.bias_output
        self.output_layer_output = self.sigmoid(self.output_layer_input)
        print(self.output_layer_output )

        return self.output_layer_output

    def backward(self, x, y, output):
        # Backpropagation
        print(x)
        print(y)
        error_output = y - output
        delta_output = error_output * self.sigmoid_derivative(output)

        error_hidden = delta_output.dot(self.weights_output.T)
        delta_hidden = error_hidden * self.sigmoid_derivative(self.hidden_layer_output)

        # Update weights and biases
        self.weights_output += self.hidden_layer_output.T.dot(delta_output)
        self.bias_output += np.sum(delta_output, axis=0, keepdims=True)

        self.weights_hidden += x.T.dot(delta_hidden)
        self.bias_hidden += np.sum(delta_hidden, axis=0, keepdims=True)

    def train(self, x, y, epochs):
        for _ in range(epochs):
            output = self.feedforward(x)
            self.backward(x, y, output)

if __name__ == "__main__":
    # Sample dataset (XOR problem)


    # Create and train the MLP
    mlp = NeuralNetwork(input_size=2, hidden_size=4, output_size=1)
    mlp.train(x[0], Y[0], epochs=10000)

    # Test the trained model
    for x in X:
        result = mlp.feedforward(x)
        print(f"Input: {x}, Predicted Output: {result}")
```
